```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
# Load the dataset
dataframe = pd.read_csv('news.csv')
```

```python
# Counting number of rows and columns in the data
print('-' * 40)
print('Data Shape:', dataframe.shape)
print('-' * 40)

# Getting a hang of the data in each column and their names
print('Sample Data:')
print(dataframe.head())
print('-' * 40)

# Looking for any places where data has NaN values
print('Null Value Count in "text" Column:', dataframe['text'].isna().sum())
print('-' * 40)

# Dropping all rows where "text" column is NaN
dataframe.dropna(subset=['text'], inplace=True)
print('Data Shape after Dropping NaNs:', dataframe.shape)
print('-' * 40)
```

```python
# Split dataset into features (x) and labels (y)
x = dataframe['text']
y = dataframe['label']

# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Initialize TF-IDF vectorizer
tfvect = TfidfVectorizer(stop_words='english', max_df=0.7)

# Transform training and testing data using TF-IDF
tfid_x_train = tfvect.fit_transform(x_train)
tfid_x_test = tfvect.transform(x_test)
```

```python
# Initialize Passive Aggressive Classifier
classifier = PassiveAggressiveClassifier(max_iter=50)
classifier.fit(tfid_x_train, y_train)
```

```python
# Predict on test data
y_pred = classifier.predict(tfid_x_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print('-' * 40)
print(f'Accuracy: {round(accuracy * 100, 2)}%')
print('-' * 40)
```

```python
# Print classification report
print('-' * 40)
from sklearn.metrics import classification_report
print('Classification Report:')
print('-' * 40)
print(classification_report(y_test, y_pred))
print('-' * 40)
```

```python
# Confusion Matrix Visualization
conf_matrix = confusion_matrix(y_test, y_pred, labels=['REAL', 'FAKE'])
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['REAL', 'FAKE'], yticklabels=['REAL', 'FAKE'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Visualization: Class Distribution Bar Plot
plt.figure(figsize=(8, 6))
sns.countplot(x='label', data=dataframe)
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization: Class Proportion Pie Chart
class_counts = dataframe['label'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Class Proportion')
plt.show()
```

```python
# Save the model to a file
with open('model.pkl', 'wb') as model_file:
    pickle.dump(classifier, model_file)

# Load the model from disk
with open('model.pkl', 'rb') as model_file:
    loaded_model = pickle.load(model_file)
```

```python
def fake_news_det(news):
    input_data = [news]
    vectorized_input_data = tfvect.transform(input_data)
    prediction = loaded_model.predict(vectorized_input_data)
    return prediction[0]

# Test the function with sample news articles
news1 = "President Barack Obama has been campaigning hard for the woman..."
news2 = "U.S. Secretary of State John F. Kerry said Monday that he will stop in Paris..."
news3 = "hello."
print("Prediction for news1:", fake_news_det(news1))
print("Prediction for news2:", fake_news_det(news2))
print("Prediction for news3:", fake_news_det(news3))
```
